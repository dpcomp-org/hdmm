```python
import numpy as np
```

```python
B = np.random.rand(3,3)
C = np.random.rand(4,4)
A = np.kron(B,C)
A1 = np.rollaxis(A.reshape(3,4,3,4), 2, 1).reshape(3*3,4*4)
```

```python
U, S, V = np.linalg.svd(A1, False)
print np.allclose(U.dot(np.diag(S)).dot(V), A1)
print np.allclose(np.outer(U[:,0],V[0]) * S[0],A1)
```
